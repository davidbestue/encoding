```python
path_L1 = 'C:\\Users\\David\\Desktop\\KI_Desktop\\data_reconstructions\\IEM\\wm_l1.xlsx'
xl = pd.ExcelFile(path_L1)
sheet_names = xl.sheet_names
```

```python
ROIs_ch_means = []

for i in range(len(sheet_names)):
    WMx = pd.read_excel(path_L1, sheet_name=sheet_names[i])
    mean_ROI = WMx.mean(axis=0)[1:].values
    ROIs_ch_means.append(mean_ROI)
```

```python
##Mean Weight of each subject
plt.figure()
for s in [0,4,8,12,16,20]:
    plt.plot(np.array(ROIs_ch_means[s:s+4]).mean(axis=0))
    plt.xticks([0,9,18,27,35], [0,90, 180, 270, 350])
    plt.ylabel('weights')
    plt.xlabel('angle')

plt.show()
    
### Mean Weights
plt.figure()
plt.plot(np.array(ROIs_ch_means).mean(axis=0))
plt.xticks([0,9,18,27,35], [0,90, 180, 270, 350])
plt.ylabel('weights')
plt.xlabel('angle')
plt.show()
```
